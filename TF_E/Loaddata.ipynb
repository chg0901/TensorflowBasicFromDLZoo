```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777) # for reproducibility
```

```python
xy = np.loadtxt('data.csv',delimiter=',',dtype=np.float32)
```

```python
x_data = xy[:, 0:-1]
y_data = xy[:,[-1]]  ##### error

print(x_data.shape,'\n', x_data,'\n',len(x_data))
print(y_data.shape,'\n', y_data,'\n',len(y_data))
```

```python
# placehiolders for a tensor that will be always fed
X = tf.placeholder(tf.float32,shape=[None,3])
Y = tf.placeholder(tf.float32,shape=[None,1])

W = tf.Variable(tf.random_normal([3,1]),name="weight")
b = tf.Variable(tf.random_normal([1]),name="bias")

# Hypothsis 
hypothsis = tf.matmul(X,W)+b
# simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothsis - Y))

# minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
# Launch the graph in a Session
sess = tf.Session()
# Initialize global variables in the graph
sess.run(tf.global_variables_initializer())
# Set up feed_dict variables inside the loop
for step in range(2001):
    cost_val,hy_val,_ = sess.run([cost,hypothsis,train],
                feed_dict= {X:x_data,Y:y_data})
    if step%10==0:
        print(step,"Cost:",sess.run(tf.transpose(cost_val)),"\nPrediction:\n",hy_val)
# Ask my Score        
print("your score will be",
      sess.run(hypothsis,feed_dict={X:[[100,70,101]]}))
print("other scores will be",
     sess.run(hypothsis,feed_dict={X:[[60,70,100],[90,100,80]]}))
```
