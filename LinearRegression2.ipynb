```python
import tensorflow as tf

# x and y data
# x_train = [1,2,3]
# y_train = [1,2,3]

# use placeholder to set x and y data
## placeholders for a tensorn that will be always fed using feed_dict
X = tf.placeholder(tf.float32,shape=[None])
Y = tf.placeholder(tf.float32,shape=[None])

# W weight and b bias
W = tf.Variable(tf.random_normal([1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')

# our hypothsis
hypothsis = X * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothsis - Y))

# GradientDescent to Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

# launch the graph in a session
sess = tf.Session()

# inintializes global variables in the graph
sess.run(tf.global_variables_initializer())

print("step,sess.run(cost),sess.run(W),sess.run(b)")
# fit the line
for step in range(2001):
    cost_val,W_val,b_val,_ = sess.run([cost,W,b,train],feed_dict = {X:[1,2,3,4,5],Y:[2.1,3.1,4.1,5.1,6.1]})
    if step % 20 == 0:
        print(step,cost_val,W_val,b_val)
```
